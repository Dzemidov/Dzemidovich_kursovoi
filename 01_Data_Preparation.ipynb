```python
import pandas as pd
```

```python
#Загрузка основного файла
sale_max = pd.read_csv('/home/roman/PycharmProjects/Dzemidovich_kursovoi/data/export_orders.csv')
```

```python
sale_max.describe()
```

```python
sale_max
```

```python
#Удаление нерелевантных столбцов
sale_max = sale_max.drop(['Коммент. клиента','Ваш ответ', 'Заказ/Счет', 'Поставщик', 'Менеджер', 'Последнее изменение'], axis=1)
```

```python
#Фильтрация строк:
sale_max = sale_max[sale_max['Клиент'].notna()]
```

```python
sale_max = sale_max[sale_max['Статус'].notna() & (sale_max['Статус'] == 'Заказ выдан')]
```

```python
sale_max
```

```python
sale_max['Дата'] = pd.to_datetime(sale_max['Дата'], format='%d.%m.%Y %H:%M:%S', errors='coerce')
```

```python
# Очистка числовых столбцов
numeric_cols = ['Кол-во заказано', 'Цена, BYN', 'Цена поставщика', 'Сумма, BYN', 'Сумма закупки', 'Итоговое кол-во']
for col in numeric_cols:
    sale_max[col] = pd.to_numeric(sale_max[col], errors='coerce')
```

```python
# Обработка пропущенных значений
sale_max = sale_max.dropna()
```

```python
# Инженерия признаков
sale_max['Profit'] = sale_max['Сумма, BYN'] - sale_max['Сумма закупки']
sale_max['Profit Margin'] = sale_max['Profit'] / sale_max['Сумма, BYN']
sale_max['Month'] = sale_max['Дата'].dt.to_period('M')
```

```python
# Статистические показатели
stats = {
    'Общая выручка (BYN)': sale_max['Сумма, BYN'].sum(),
    'Общая прибыль (BYN)': sale_max['Profit'].sum(),
    'Средняя прибыль на заказ (BYN)': sale_max['Profit'].mean(),
    'Стандартное отклонение прибыли на заказ (BYN)': sale_max['Profit'].std(),
    'Медиана прибыли на заказ (BYN)': sale_max['Profit'].median(),
    'Минимальная прибыль (BYN)': sale_max['Profit'].min(),
    'Максимальная прибыль (BYN)': sale_max['Profit'].max(),
}
stats
```
